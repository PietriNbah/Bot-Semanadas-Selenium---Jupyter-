```python
import time
import pyautogui
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

#Definindo variaveis
email = input("Qual o seu RA?") + "@eniac.edu.br"
senha = input("Qual a sua senha?")
#curso = input("Qual o seu curso?(Por favor responda com apenas o numero, exemplo do segundo ano: 2)")
#cur_tec = input("Qual  seu curso tecnico?")
n_semanada = input("Qual o numero das semanadas que voce deseja ver?(Digite apenas o numero, exemplo: 1)")

#Variaveis semanada:
semanada_algebra = "Semanada " + n_semanada + " - Álgebra"
semanada_geometria = "Semanada " + n_semanada + " - Geometria"
semanada_estatistica = "Semanada " + n_semanada + " - Estatística"
semanada_fisica = "Semanada " + n_semanada + " - Física"
semanada_quimica = "Semanada " + n_semanada + " - Química"
semanada_biologia = "Semanada " + n_semanada + " - Biologia"
semanada_gramatica = "Semanada " + n_semanada + " - Gramática"
semanada_literatura = "Semanada " + n_semanada + " - Literatura"
semanada_ingles = "Semanada " + n_semanada + " - Inglês"
semanada_historia = "Semanada " + n_semanada + " - História"
semanada_geografia = "Semanada " + n_semanada + " - Geografia"
semanada_filosofia = "Semanada " + n_semanada + " - Filosofia"
#semanada_tecnico = "Semanada " + n_semanada + " - Álgebra"
lista = [semanada_algebra, semanada_geometria, semanada_estatistica, semanada_fisica, semanada_quimica, semanada_biologia, semanada_gramatica, semanada_literatura, semanada_ingles, semanada_historia, semanada_geografia, semanada_filosofia]

nav = webdriver.Chrome() # entrando no portal
nav.get("https://portalacademico.eniac.edu.br/")

#login
nav.find_element(By.XPATH,'//*[@id="snap-pm-inner"]/div[1]/div[4]/div/div/div/a').click()#Gsuite
nav.find_element(By.XPATH, '//*[@id="identifierId"]').send_keys(email)
nav.find_element(By.XPATH, '//*[@id="identifierId"]').send_keys(Keys.ENTER)#RA
time.sleep(10)
nav.find_element(By.XPATH, '//*[@id="password"]/div[1]/div/div[1]/input').send_keys(senha)
nav.find_element(By.XPATH, '//*[@id="password"]/div[1]/div/div[1]/input').send_keys(Keys.ENTER)#senha
time.sleep(15)
pyautogui.press("esc") # sai da tela chata
nav.find_element(By.PARTIAL_LINK_TEXT, 'Semanadas').click() # semanadas
nav.find_element(By.XPATH, '//*[@id="section-0"]/div/nav/a/span').click() # semanadas realizadas
nav.find_element(By.XPATH, '//*[@id="section-1"]/div/nav/a[2]/span/span').click() # PROXIMA SESSAO - Semanadas realizadas

# pegando as notas

#algebra
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_algebra).click() # entra na prova
nota_algebra = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_algebra)
time.sleep(5)

#geometria
nav.find_element(By.PARTIAL-LINK_TEXT, semanada_geometria).click() # entra na prova
nota_geometria = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_geometria)
time.sleep(5)

#estatistica
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_estatistica).click() # entra na prova
nota_estatistica = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_estatistica)
time.sleep(5)

#fisica
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_fisica).click() # entra na prova
nota_fisica = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_fisica)
time.sleep(5)

#quimica
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_quimica).click() # entra na prova
nota_quimica = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_quimica)
time.sleep(5)

#biologia
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_biologia).click() # entra na prova
nota_biologia = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_biologia)
time.sleep(5)

#gramatica
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_gramatica).click() # entra na prova
nota_gramatica = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_gramatica)
time.sleep(5)

#literatura
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_literatura).click() # entra na prova
nota_literatura = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_literatura)
time.sleep(5)

#ingles
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_ingles).click() # entra na prova
nota_ingles = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_ingles)
time.sleep(5)

#historia
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_historia).click() # entra na prova
nota_historia = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_historia)
time.sleep(5)

#geografia
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_geografia).click() # entra na prova
nota_geografia = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_geografia)
time.sleep(5)

#filosofia
nav.find_element(By.PARTIAL_LINK_TEXT, semanada_filosofia).click() # entra na prova
nota_filosofia = nav.find_element(By.XPATH, '//*[@id="region-main"]/div[2]/table/tbody/tr/td[2]').text #pega nota
nav.find_element(By.LINK_TEXT, "Semanadas Realizadas").click() #voltas semanadas relaizadas
print(nota_filosofia)
time.sleep(5)
```
